***#Password_checker(check strength of passwords)***


```python
import re  # re-regular expression


def pass_strength(password):
    score = 0
    feedback = []

    # this condition checks the length
    if len(password) >= 8:
        score += 1  # means-true
    else:
        feedback.append("Min. length should be 8 characters")

    # this is for lowercase
    if re.search(r"[a-z]", password):
        score += 1
    else:
        feedback.append("Password must carry one lowercase")

    # this is for uppercase
    if re.search(r"[A-Z]", password):
        score += 1
    else:
        feedback.append("Password must have one uppercase")

    # this will check numbers
    if re.search(r"\d", password):
        score += 1
    else:
        feedback.append("Password must include a number")

    # this will check the symbol
    if re.search(r"[!@#$%^&*~`,.;:\"" "]", password):
        score += 1
    else:
        feedback.append("Password must have one symbol")


    strength = "Very Weak"

    if score == 5:
        strength = "Strong"
    if score == 4:
        strength = "Medium"  
    if score == 3:
        strength = "Weak"
    if score == 2:
        strength = "Very weak"
    if score == 1:
        strength = "NA"

    # The function ENDS here
    return {"strength": strength, "feedback": feedback, "score": score}



result = input("Enter or Create Password: ")
test_password = pass_strength(result)



print("Password Strength is:", test_password["strength"] ,test_password["feedback"])
#print("Suggestions:", test_password["feedback"])
#this will create the problem to encrypt password easily rather the password satisfy all the coditions.
```

#to overcome from that problem (password like:-'Password@123' i.e.,also weak password),there is 'zxcvbn':Dropbox created a tool called zxcvbn that checks passwords against common patterns, dictionary words, names, and keyboard patterns.


```python
from zxcvbn import zxcvbn

def evaluate_password(password):
    # You can also pass a user_inputs list (e.g., [username, email]) to check for duplicates
    analysis = zxcvbn(password)
    
  
    score = analysis['score']
    strength_levels = {0: "", 1: "Very Weak", 2: "Weak", 3: "Good", 4: "Strong",5:"very strong"}
    
    print(f"Password: {password}")
    print(f"Strength Level: {strength_levels[score]}")
    print(f"Estimated cracks needed: {analysis['crack_times_display']['offline_fast_hashing_1e10_per_second']}")
    
    # Provide feedback if available
    feedback = analysis['feedback']
    if feedback['warning']:
        print(f"Warning: {feedback['warning']}")
    if feedback['suggestions']:
        print("Suggestions:")
        for suggestion in feedback['suggestions']:
            print(f" - {suggestion}")

# Test with a weak but complex-looking password
evaluate_password("Password123!") 
print("-" * 30)
# Test with a long, uncommon phrase (much safer)
evaluate_password("correct-horse-battery-staple")
```
